```python
import pandas as pd

# Create a sample pandas DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'London', 'Tokyo']
})
print(df)
```

```python
def modify_df(df, new_name=None, age_increment=0, new_city=None):
    """Modify the dataframe by updating Name, Age, or City columns."""
    modified = df.copy()
    if new_name:
        modified['Name'] = new_name
    if age_increment:
        modified['Age'] = modified['Age'] + age_increment
    if new_city:
        modified['City'] = new_city
    return modified

# Example usage
modified_df = modify_df(df, age_increment=1, new_city='Paris')
print(modified_df)
```
